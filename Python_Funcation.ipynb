# User defined funcation

```python
#python program for funcation that takes a list of numbers and return the list of their sequares
def square_list(numbers):
    return [x**2 for x in numbers]
print(square_list([1, 2, 3, 4]))  
```

```python
#creat a funcation that takes a string and counts the number of vowels
def count_vowels(text):
    vowels = 'aeiouAEIOU'
    return sum(1 for char in text if char in vowels)
print(count_vowels("Hello World")) 
```

```python
#Define a funcation that accepts a number and returns true if it's an armstrong numbeer
def is_armstrong(num):
    digits = [int(d) for d in str(num)]
    power = len(digits)
    return num == sum(d**power for d in digits)
print(is_armstrong(153))
```

```python
#write a funcation that accepts a number and returns names that start with the letter "s"
def names_starting_with_s(names):
    return [name for name in names if name.startswith('S')]
print(names_starting_with_s(["Sam", "Alex", "Sara", "John"]))  
```

```python
#create a funcation that accepts two numbers and return their HCF(Highest common factor)
def find_hcf(a, b):
    while b:
        a, b = b, a % b
    return a
print(find_hcf(48, 18))  
```

# Lambda Function

```python
#Program to check if a number is divisible by 5
is_divisible_by_5 = lambda x: x % 5 == 0
print(is_divisible_by_5(25))
```

```python
#Use a lambda to convert temperature from celsius to fahrenheit
c_to_f = lambda c: (c * 9/5) + 32
print(c_to_f(0))
```

```python
#create a lambda funcation that return the cube of a number
cube = lambda x: x ** 3
print(cube(3))
```

```python
#write a lambda funcation to add of two string
add_strings = lambda s1, s2: s1 + s2
print(add_strings("Hello, ", "World!"))  
```

```python
# Use a lambda to check if the length of a string is greater than 8
is_long = lambda s: len(s) > 8
print(is_long("OpenAI GPT"))
```

# Filter Funcation

```python
# Use filter() to get all strings with length greater than 4 from a list
words = ["apple", "bat", "banana", "dog", "elephant"]
result = list(filter(lambda x: len(x) > 4, words))
print(result)
```

```python
# Use filter()to extract numbers greater than 100 from a list
numbers = [50, 120, 99, 101, 200, 30]
result = list(filter(lambda x: x > 100, numbers))
print(result)
```

```python
#write a program to filter all odd numbers from a list using a lambda
numbers = [10, 15, 22, 33, 42, 55]
result = list(filter(lambda x: x % 2 != 0, numbers))
print(result)  
```

```python
#Use filtere() to find names that end with the letter 'a'
names = ["Sita", "Ram", "Gita", "Hari", "Radha"]
result = list(filter(lambda x: x.endswith('a'), names))
print(result) 
```

```python
# filter out all elements that are not integer from a mixed list
mixed = [1, "hello", 3.14, 42, True, 100, "world", None]
result = list(filter(lambda x: isinstance(x, int) and not isinstance(x, bool), mixed))
print(result)
```

# Map Funcation

```python
#Use map() to convert a list of integers into their squares
nums = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x**2, nums))
print(squares)
```

```python
#Apply map() to convert a list of strings to uppercase
words = ['apple', 'banana', 'cherry']
uppercase_words = list(map(str.upper, words))
print(uppercase_words)
```

```python
#Use map() with lambda to subtract 2 from every number in a list
nums = [5, 10, 15, 20]
subtracted = list(map(lambda x: x - 2, nums))
print(subtracted)
```

```python
#use map() to get the length of each word in a sentence
sentence = "Map functions are useful"
words = sentence.split()
lengths = list(map(len, words))
print(lengths)
```

```python
# Apply map() to multiplty each element of a list by its index
nums = [4, 5, 6, 7]
multiplied = list(map(lambda x: x[0] * x[1], enumerate(nums)))
print(multiplied)
```
