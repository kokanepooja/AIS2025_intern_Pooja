```python
# Python program factorial of a number using a for loop
num = int(input("Enter a number: "))
factorial = 1
for i in range(1, num + 1):
    factorial *= i
print(f"Factorial of {num} is {factorial}")
```

# while loop

```python
#Python program from 1 to 10 using while loop
i = 1
while i <= 10:
    print(i)
    i += 1
```

```python
#Python program for sum number until the user enter 0
total = 0
num = int(input("Enter a number (0 to stop): "))
while num != 0:
    total += num
    num = int(input("Enter a number (0 to stop): "))
print(f"Total sum: {total}")
```

```python
# Reverse a given number using while
num = int(input("Enter a number: "))
rev = 0
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10
print(f"Reversed number: {rev}")
```

```python
#Python program for count digits of a number using while
num = abs(int(input("Enter a number: ")))
count = 0
if num == 0:
    count = 1
else:
    while num > 0:
        num //= 10
        count += 1
print(f"Number of digits: {count}")
```

```python
#Python program for print fibonacci series up to n terms using while
n = int(input("Enter number of terms: "))
a, b = 0, 1
count = 0
while count < n:
    print(a)
    a, b = b, a + b
    count += 1
```

```python
# while loop
import time
response = "Sure, let me help you with that."
i = 0
while i < len(response):
    print(response[i], end='', flush=True)
    time.sleep(0.05)  
    i += 1
```

```python
import random
price = 25000
target = 26000

while price < target:
    print(f"Current BTC Price: ${price}")
    price += random.randint(100, 300) 6
    time.sleep(1)
print("Target reached!")
```

```python
goal = 10000
steps = 0

while steps < goal:
    steps += 500 
    print(f"Steps walked: {steps}")
    time.sleep(0.5)
print("Daily step goal reached!")
```

```python
import random
attempts = 0
solved = False
while not solved:
    prediction = random.choice(["correct", "incorrect"])
    attempts += 1
    print(f"Attempt {attempts}: {prediction}")
    if prediction == "correct":
        solved = True
print(f"Captcha solved in {attempts} attempts!")
```

```python
canvas_limit = 1000
pixels_drawn = 0
while pixels_drawn < canvas_limit:
    shape_size = random.randint(10, 100)
    pixels_drawn += shape_size
    print(f"Drew a shape of {shape_size} pixels. Total: {pixels_drawn}/{canvas_limit}")
    time.sleep(0.3)
print("Canvas is full")
```

# control statement

```python
# Print numbers 1 to 10, but stop if number is 5 (using break)
for i in range(1, 11):
    if i == 5:
        break
    print(i)
```

```python
#Print numbers 1 to 10, but skip 5 (using continue)
for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```

```python
# Use pass statement inside a loop and explain its purpose
for i in range(1, 6):
    if i == 3:
        pass 
    print(f"Number is: {i}")
```

```python
# Break statement
secret = 7
while True:
    guess = int(input("Guess the number (1–10): "))
    if guess == secret:
        print("🎉 Correct! You guessed it.")
        break
    else:
        print("❌ Wrong. Try again.")
```

```python
# continue statement
message = "Meet at 9 PM. Code RED. Avoid 5th Ave!"
for char in message:
    if char in "aeiouAEIOU":
        continue  # skip vowels
    print(char, end="")
```

```python
# pass statement
sequence = "ATCGGCTXAG"
for base in sequence:
    if base not in "ATCG":
        pass  # placeholder for error handling
    else:
        print(f"Valid base: {base}")
```

```python
# Break statement
shopping_list = ["milk", "bread", "butter", "coffee", "eggs"]
item_to_find = "coffee"

for item in shopping_list:
    if item == item_to_find:
        print(f"✅ {item} is in your list.")
        break
```
