# Set

# Introducation to set:
sets are unordered collectations of unique elements. Useful for memebership tests and eliminating duplicates.
# key points
-Elements must be immutable(e.g.,numbers,string,tuples)
-No duplicate values allowed.
-Supports set operations like union (|),intersecation (&),difference (-),symmetric difference (^).
-Methods include add(),remove(),discard(),pop(),clear().
-Unordered-no indexing or slicing


```python
data={110,10,20,230,240}
print(data)
print(type(data))
```

```python
# creat set
fruit={"apple","banana","cherry","apple"}
print(fruit)
```

```python
# Add element
fruit.add("orange")
print("After adding orange:",fruit)
```

```python
# Remove an element
fruit.remove("banana")
print("After removing banana:",fruit)
```

```python
# check memebership
print("Is 'cherry' in the set?","cherry" in fruit)
```

```python
# set operations
set1={1,2,3}
set2={3,4,5}
```

```python
# Union(All elements from both set)
print("Union:",set1 | set2)
```

```python
# intersecation (common elements)
print("Intersecation:",set1 & set2)
```

```python
# Difference (Element in set1 but not in set2)
print("Difference:",set1-set2)
```

```python
#Check if set1 is a subset of set2
print("Is set1 a subset of set2?",set1.issubset(set2))
```

```python
# check if set2 is a superset of set1
print("Is set2 a subset of set1?",set2.issubset(set1))
```

```python
# check if sets are disjiont(No common elements)
c={6,7}
print("Is set1 disjoint with c?",set1.isdisjoint(c))
```
