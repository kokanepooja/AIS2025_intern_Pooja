# Dictonary

# Introducation to dictonary(dict)
Dictionaries store data in key-value pairs.key must be unique and immutable.
# key Points
-Access value using keys,not indexes.
-Dynamic-can add or remove key-value pairs anytime.
-Support methods like keys(),values(),items(),get(),pop(),update().
-Efficient for fast lookup and retrieval of data by keys.
-Nested dictionaries are also possible.

```python
player={18:"virat",45:"R0hit"}
print(player)
print(type(player))
player[18]
```

```python
# create a dictionary
student={"name":"Pooja","age":22,"course":"statistics"}
print("Student dictionary:",student)
```

```python
#Access a value using a key
print("Name:",student["name"])
```

```python
#Add a new key-value pair
student["grade"]=21
print("Updated age:",student)
```

```python
#update an existing value
student["age"]=21
print("updated age:",student)
```

```python
#Remove a key-value pair
student.pop("course")
print("After removing course:",student)
```

```python
# Get all keys,values and items
print("keys:",student.keys())
print("value:",student.values())
print("Item:",student.items())
```

```python
#use get() to access values safely
print("course:",student.get("course","Not found"))
```

```python
# Nested dictionaries
student={"student1":{"name":"Pooja","age":22},"student2":{"name":"Yash","age":22}}
```
