# Tuple

# Introducation to tuples(tuple)
Tuple are ordered collections likes list,but they are immutable.once created,their values cannot be changed.
# Key Points
-Faster than lists for read-only data.
-Useful for fixed collections (e.g.,cordinates)
-Supports indexing,slicing and some built-in funcations.
-Tuples can also contain mixed data types.
-can be used as keys in dictionaries (if they contain only immutable elements).

```python
emtuple=(18,"pooja",25.5,"neha",30.5)
print(emtuple)
print(type(emtuple))
```

```python
person=("Alice",25,"Engineer")
#Indexing
print("Name:",person[0])
#slicing
print("Age and profession:",person[1:])
#Length
print("Length of tuple:",len(person))
#Iterartion
print("Tuple elements:")
forfor item in person:
print(item)
#concatenation
extra_info=("New York",)
full_info=person + extra_info
print("extended tuple:",full_info)
#repetition
print("Repeated tuple:",person*2)
#membership test
print("Is 'Alice in tuple?","Alice" in person)
```
