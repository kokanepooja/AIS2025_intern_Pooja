# List

# Introducation to lists (list)
Lists are ordered collection of items (of any type).lists are mutable,meaning elements can be changed.
# Key Points
-Elements can be accessed by index.
-Allows mixed data types.
-Supports method like append(), insert() ,remove() ,pop() ,sort(),etc
-Slicing can extract parts of a list
-lists can be nested (i.e.,lists within lists)

```python
list1=[18,"pooja",25.5,"neha",30.5]
print(list1)
print(type(list1))
```

# Indexing in list

```python
-Elements are accessed by their index number.
-Index starts from '0'for the first element.
-Negative indexing starts from'-1'(last element)
```

```python
print(list1[0])
print(list1[-1])
```

# Slicing a list

We can get a subset of the list using 'list[start:end]'

```python
print(list[1:3])
```

# Methods

```python
grades=["A","B","A","C","A","B"]
print(grades.index("C"))
print(grades.count("A"))
grades.sort()
print(grades)
grades.reverse()
print(grades)
```
