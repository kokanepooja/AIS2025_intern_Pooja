# String

# Introducation to string (str)
string are used to store text data in Python.They are sequence of Unicode character enclosed in single or double quotes.
# key Points
-string are immutable(cannot be changed after creation)
-can be accessed using indexing and slicing.
-Supports string methods like upper(), lower(), find(), replace(),etc
-Can be concatented using '+' and repeated using '*'
-Multiline strings can be defined using triple quotes.

```python
friend="Rukii"
print(friend)
print(type(friend))
```

 # Indexing in string

```python
indexing refers to acessing indivisual character using their position(index):
```

```python
text="Python"
print(text[0])
print(text[-1])
```

# Slicing in string

Slicing extracts a substring:

```python
print(text[0:3])
print(text[::2])
```

# Common method fuctions

```python
# 1 upper(),lower(),title()
msg="Hello"
print(msg.upper())
print(msg.title())
# 2 strip(),lstrip(),rstrip()
data="Hello"
print(data.strip())
print(data.lstrip())
# 3 find(),index()
sentence="find the letter"
print(sentence.find('e'))
# 4 replace()
phrase="I love Physics"
print(phrase.replace("Physics","Statistics"))
```
